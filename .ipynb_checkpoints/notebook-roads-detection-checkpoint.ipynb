```python
import cv2
import numpy as np

from matplotlib import pyplot as plt
```

```python
IMAGE_URL = "assets/roads-simple-2.jpeg"
```

```python
im = cv2.imread(IMAGE_URL)
```

```python
plt.imshow(im)
```

```python
im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
plt.imshow(im_gray)
```

```python
(thresh, im_black_and_white) = cv2.threshold(im_gray, 127, 255, cv2.THRESH_BINARY)
plt.imshow(im_black_and_white)
```

```python
image = im_black_and_white
se=cv2.getStructuringElement(cv2.MORPH_RECT , (8,8))
bg=cv2.morphologyEx(image, cv2.MORPH_DILATE, se)
out_gray=cv2.divide(image, bg, scale=255)
out_binary=cv2.threshold(out_gray, 0, 255, cv2.THRESH_OTSU)[1] 
plt.imshow(out_binary)
```

```python
ksize = (150, 130)  
blurred = cv2.blur(out_binary, ksize, cv2.BORDER_DEFAULT) 
blurred_clean = cv2.threshold(blurred, 0, 255, cv2.THRESH_OTSU)[1]

plt.imshow(blurred_clean)
```

```python
blurred_clean_resized = cv2.resize(blurred_clean, dsize=(21, 21), interpolation=cv2.INTER_CUBIC)
plt.imshow(blurred_clean_resized)
```
